### 3.BASED ON CITY

### 4.BASED ON STATE

```python
 data['State'].unique()
```

```python
data['State'].value_counts()
```

```python
st=data.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()
st.head()
```

## PROFIT WISE ANALYSIS IN STATES

```python
st.sort_values('Profit', ascending=False, inplace=True)
```

```python
st.head(10)
```

```python
st[['Profit']].plot(kind ='bar', figsize=(20,8))
plt.xlabel('Name of States')
plt.ylabel('Profit')
plt.title('Profit vs States')
plt.legend()
```

HIGHEST PROFIT GIVING STATE IS "VERMONT"

LOWEST PROFIT GIVING STATE IS "OHIO"

## SALE WISE ANALYSIS IN STATE

```python
st[['Sales']].plot.pie(subplots=True,autopct='%1.2f%%' ,figsize=(30,20), labels=st.index)
```

```python
st1= st.sort_values('Sales', ascending=False)
```

```python
st1[['Sales']].head(30).plot(kind='bar',color='orange',figsize=(20,8))
plt.xlabel('Sates', fontsize=15)
plt.ylabel('Sales', fontsize=15)
plt.title('Plot of Top 30 Sales State', fontsize=20)
```

HIGHEST SALE GIVING STATE IS WYOMING

```python
st1[['Sales']].tail(30).plot(kind='bar',color='grey',figsize=(20,8))
plt.xlabel('Sates', fontsize=15)
plt.ylabel('Sales', fontsize=15)
plt.title('Plot of Least Sales State', fontsize=20)
```

LOWEST SALE GIVING STATE IS SOUTH DAKOTA

## DISCOUNT WISE ANALYSIS IN STATE

```python
st['Discount'].plot(kind ='bar', figsize=(20,8))
plt.xlabel('State', fontsize= 15)
plt.ylabel('Discount', fontsize= 15)
plt.title('Discount wise analysis in state ')
```

HIGHEST DISCOUNT GIVING STATE IS ILLINOIS
